(e) Plot by category

<font size = "3">

- Read "data/wdi_urban.csv"<br>
This contains data from 1980 and 2020

- Create a list with the unique values of "year"
- Create a plot with a ...
    - Histogram of "prop_urbanpopulation" <br>
    for each year
    - Set the histogram option ``` alpha = 0.5 ``` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title

What do we learn from the graph?

HINT: See the example at the end of Lecture 10


````python
# Write your own code here

wdi_urban = pd.read_csv("data/wdi_urban.csv")
list_years = pd.unique(wdi_urban["year"])

for year in list_years:
    df = wdi_urban.query("year == @year")
    plt.hist(df["prop_urbanpopulation"], alpha = 0.5)
    
plt.xlabel("Prop_urbanpopulation")
plt.ylabel("Frequency")
plt.legend(labels = list_years, 
           title = "Years")
plt.show()
````

<font size = 5>We learn from this graph that, generally speaking, the prop_urbanopulation increases from 1980 to 2020. In 1980, the data is left-skewed distributed, whereas it becomes more right-skewed distributed in 2020. </font>
